```python
import numpy as np 
import random
import matplotlib.pyplot as plt
import math
```

Constants:

```python
n = 20
x_mean = 30
x_var = 20
a = 1
b = 1
c = 3
d = 0.01
mu = a / c
tau = c

ITERATIONS = 100000
```

Gibbs sampler:

```python
j = 1
mu_list = []
tau_list = []
iter_list = []
sigma_list = []
while j <= ITERATIONS:
    mu = np.random.normal((c*d + n*tau*x_mean)/(n*tau + d), math.sqrt(1/(n*tau + d)))
    tau = np.random.gamma(a + n/2, 1/(b + (1/2)*((n-1)*x_var + n*(x_mean - mu)**2)))
    tau_list.append(tau)
    mu_list.append(mu)
    sigma_list.append(1/math.sqrt(tau))
    iter_list.append(j)
    j += 1
```

Marginal distribution for tau:

```python
plt.hist(tau_list, bins=400)
plt.show()
```

Marginal distribution for mu:

```python
plt.xlim(20,40)
plt.hist(mu_list, bins=400)
plt.show()
```

Marginal distribution for sigma:

```python
plt.xlim(0, 10)
plt.hist(sigma_list, bins=400)
plt.show()
```
